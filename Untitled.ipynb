```
#import some basic libs
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import cv2
from google.colab import files
from io import BytesIO
from PIL import Image
from numpy import linalg as LA
from scipy.spatial import distance
```

```
#upload all training images
for i in range(8):
  uploaded = files.upload()
```

```
#saving all the images
im1 = Image.open("training1.jpg")
im2 = Image.open("training2.jpg")
im3 = Image.open("training3.jpg")
im4 = Image.open("training4.jpg")
im5 = Image.open("training5.jpg")
im6 = Image.open("training6.jpg")
im7 = Image.open("training7.jpg")
im8 = Image.open("training8.jpg")

img = [0]*8
img[0]=im1
img[1]=im2
img[2]=im3
img[3]=im4
img[4]=im5
img[5]=im6
img[6]=im7
img[7]=im8
```

```
#plot all the training images and Stacked together each training image to form a column vector ùëÖi0.
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

R = [0]*8

for i in range(8):
  plt.imshow(img[i], cmap='gray')
  plt.show()

  im = np.asarray(img[i])
  R[i] = im.flatten('C')
  R[i] = R[i].reshape(R[i].size,1)
```

```
#Calculate the mean face

mean = np.ndarray(shape=(img[0].size[0]*img[0].size[1],1))

for i in range(8):
  mean += R[i]

mean = np.divide(mean,8)

plt.imshow(mean.reshape(img[0].size[1],img[0].size[0]), cmap='gray')
plt.show()
```

```
#subtract the mean face for each training image and put them together
Rnew = [0]*8
for i in range(8):
  Rnew[i] = np.subtract(R[i],mean)
  plt.imshow(Rnew[i].reshape(img[0].size[1],img[0].size[0]), cmap='gray')
  plt.show()

A = np.concatenate((Rnew[0], Rnew[1], Rnew[2], Rnew[3], Rnew[4], Rnew[5], Rnew[6], Rnew[7]), axis=1)
```

```
#Form covariance matrix and plot eigenfaces
L = np.dot(A.transpose(),A)
eigenvalues, eigenvectors = np.linalg.eig(L)
U = np.dot(A,eigenvectors)

Uplot = np.divide(np.dot(np.subtract(U,U.min()),255),np.subtract(U.max(),U.min()))

for i in range(8):
  f1 = Uplot[:,i]
  plt.imshow(f1.reshape(img[0].size[1],img[0].size[0]),cmap='gray')
  plt.show()
```

```
#Compute eigenface coefficients
coe = [0]*8

for i in range(8):
  coe[i] = np.dot(U.transpose(),Rnew[i])
  print("training eigenface coefficients",i+1,": \n",coe[i])


```

```
def Edist(a,b):
  sum = 0.0
  for i in range(8):
    sum += (a[i]-b[i])**2
  sum = np.sqrt(sum)
  return sum
```

```
#Compute distance and find the minimal.
closet = [1]*5
smallest = 0.0

for i in range(5):
  smallest = Edist(coet[i],coe[0])
  for j in range(8):
    value = Edist(coet[i],coe[j]);
    if value<smallest:
      smallest = value
      closet[i] = j+1

for i in range(5):
  print("Testing image ",i+1)
  plt.imshow(img2[i], cmap='gray')
  plt.show()

  print("It belongs to training image ",closet[i])

  plt.imshow(img[closet[i]-1], cmap='gray')
  plt.show()
  print("\n\n")
```
